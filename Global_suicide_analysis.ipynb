```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/master.csv')
```

```python
# Data preprocessing
# You may need to clean the dataset, handle missing values, and perform necessary transformations
```

Analyze and Visualization

```python
# 1. Basic statistics
total_suicides = data['suicides_no'].sum()
mean_suicides = data['suicides_no'].mean()
```

```python
print("Total suicides:", total_suicides)
print("Mean suicides per year:", mean_suicides)
```

```python
# 2. Suicide rates by year
suicides_by_year = data.groupby('year')['suicides_no'].sum()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(suicides_by_year.index, suicides_by_year.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of suicides')
plt.title('Global Suicides by Year')
plt.grid(True)
plt.show()
```

```python
# 3. Suicide rates by country
suicides_by_country = data.groupby('country')['suicides_no'].sum().sort_values(ascending=False)
```

```python
# Bar plot of suicide count by country
plt.figure(figsize=(18, 6))
sns.barplot(x='country', y='suicides_no', data=data)
plt.title('Suicide Count by Country')
plt.xlabel('Country')
plt.ylabel('Suicide Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar Plot of top 10 Countries with the Highest Number of Suicides
plt.figure(figsize=(12, 8))
suicides_by_country[:10].plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of suicides')
plt.title('Top 10 Countries with the Highest Number of Suicides')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.set(style="whitegrid")
```

```python
# Visualization 1: Bar plot of suicide count by gender
sns.catplot(x="sex", y="suicides_no", data=data, kind="bar", ci=None)
plt.title('Suicide Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Suicide Count')
plt.show()
```

```python
# Visualization 2: Box plot of suicide count by age group
plt.figure(figsize=(10, 6))
sns.barplot(x='age', y='suicides_no', data=data)
plt.title('Suicide Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Suicide Count')
plt.show()
```

```python
# Machine Learning: Predicting suicide rates based on age and GDP
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Machine Learning - Predicting suicide rates
# Select features and target variable
features = ['year', 'sex', 'age']
target = 'suicides_no'
X = data[features]
y = data[target]
```

```python
# Convert categorical variables to numerical using one-hot encoding
X_encoded = pd.get_dummies(X)

# Split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# Train the Decision Tree classifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```
